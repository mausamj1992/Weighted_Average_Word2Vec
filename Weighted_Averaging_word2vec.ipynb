```python
embed_len = 0
with open("karate.embeddings") as f:
    for line in islice(f, 1, None):
        arr = line.split()
        embed_len = len(arr) - 1
        break
```

```python
deg = []
kr = []
deg_sum = 0
with open("karate.adjlist") as f:
    for line in f:
        arr = line.split()
        deg.append(len(arr)-1)
        
deg_sum = sum(deg)
print deg
print deg_sum, "\n"


karate_rep = []
        
for i in range(0,10):
    karate_rep.append(0)
    kr.append(0)
    
with open("karate.embeddings") as f:
    for line in islice(f, 1, None):
        arr = line.split()
        ver = int(arr[0])    # which vertex under consideration
        #print deg[ver-1]
        arr = arr[1:]   # removing vertex label from representation vector
        #print type(deg[0])
        #print type(float(arr[0]))
        for i in range(0, len(arr)):
            karate_rep[i] = karate_rep[i] + (deg[ver-1] * float(arr[i]))
            kr[i] = kr[i] + float(arr[i])
        #arr = deg[ver-1] * float(arr)

print kr
print "\nSum of all vectors : \n"
print karate_rep

for i in range(0,embed_len):
    karate_rep[i] = karate_rep[i] / deg_sum
print "\nAfter dividing by total degree of the graph : \n"
print karate_rep
```

```python
### Calculating Euclidean distance between GRAPH VECTORS ###

dkf = 0
dks = 0
dfs = 0
dist_karate_first = 0
dist_karate_second = 0
dist_first_second = 0
for i in range(0,embed_len):
    dist_karate_first = dist_karate_first + ((karate_rep[i] - graph_rep1[i]) * (karate_rep[i] - graph_rep1[i]))
    dist_karate_second = dist_karate_second + ((karate_rep[i] - graph_rep2[i]) * (karate_rep[i] - graph_rep2[i]))
    dist_first_second = dist_first_second + ((graph_rep1[i] - graph_rep2[i]) * (graph_rep1[i] - graph_rep2[i]))
    
    dkf = dkf + ((kr[i] - gr1[i])*(kr[i] - gr1[i]))
    dks = dks + ((kr[i] - gr2[i])*(kr[i] - gr2[i]))
    dfs = dfs + ((gr1[i] - gr2[i])*(gr1[i] - gr2[i]))
    
print "\n--- SIMPLE AVERAGE ---"
print "\nDistance between KARATE and first graph : ", dkf
print "\nDistance between KARATE and second graph : ", dks
print "\nDistance between first and second graph : ", dfs

print "\n\n--- WEIGHTED AVERAGE ---"
print "\nDistance between KARATE and first graph : ", dist_karate_first
print "\nDistance between KARATE and second graph : ", dist_karate_second
print "\nDistance between first and second graph : ", dist_first_second
```
